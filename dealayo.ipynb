```python
for data in data_collection:
    img_src = data.find("img",src = True)["src"]
    product_name = data.find("h2",class_ = "product-name").text.strip()
    special_price = soup.find("p", class_ = "special-price").find("span",class_= "price").text.strip()
    old_price = soup.find("p",class_= "old-price").find("span",class_= "price").text.strip()
    print(img_src,product_name,special_price, old_price)
    
```

```python
group_menu = soup.find("ul",class_="groupmenu")
```

```python
list_href = []
```

```python
anchor = group_menu.find_all("a",href = True)
for a in anchor:
    list_href.append(a["href"])
    
```
