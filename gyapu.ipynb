```python
def filter_string(values):      #removing unnecessary sting and converting to number
    unwanted_char = "NRP,()"
    if values == None:
        return None
    else: 
        result = ''.join( c for c in values if c not in unwanted_char ).strip()       #remiving NRP and , 
        return result
```

```python
def filter_href(link):      #removing unnecessary sting and converting to number
    resutl = link.replace("https://www.okdam.com/category/",'')
    return resutl
```

```python
def strip_string(string_value):
    return string_value.strip()
```

```python
file = open('okdam.csv','w')
writer = csv.writer(file)
header = ('catagory','name','actual_price','paying_price','rating','src','vendors','img_src')
writer.writerow(header)
```

```python
for link in links_list:
    try:
        response = requests.get(link)
        soup = bs(response.content, 'html.parser')
        product_box = soup.find_all("div", class_ = "product-box")
        catagory = filter_href(link)
        for product in product_box:
    
            name = strip_string(product.find("div", class_ = "product_name-cat").text)

            p_price = product.find("p", class_ = "p-price-tag")
            paying_price = int(filter_string(p_price.find("span",class_="og-price").text))
            rating = int(filter_string(product.find("span", class_ = "rating-count").text))
            src = link
            vendors = None
            img_tag = product.find("img")
            img_src = img_tag["data-src"]

            try:
                actual_price = int(filter_string(p_price.find("del").text))
            except:
                actual_price = None
            
    
                row = (catagory,name,actual_price,paying_price,rating,src,vendors,img_src)
                writer.writerow(row)
            
    except:
        pass


        
        
```
