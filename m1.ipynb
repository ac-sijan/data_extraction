```python
def filter_string_num(Price):      #removing unnecessary sting and converting to number
    unwanted_char = "NRP,()"
    if Price == None:
        return None
    else: 
        price = ''.join( c for c in Price if c not in unwanted_char ).strip()       #remiving NRP and , 
        price = int(price)
        return price
```

```python
def filter_string(string_value):
    return string_value.strip()
```

```python
for key, value in link_dict.items():
    try:
        catagory = key
        URL = value
        response = requests.get(URL)
        soup = bs(response.content, 'html.parser')
        body = soup.body
        body_container = body.find("div",class_ = "feature-box1")
        product_inner = body_container.find_all("div",class_ = "product-inner")
        for product_in in product_inner:
            name = filter_string(product_in.find("div", class_ = "product_name-cat").text)
            p_price = product_in.find("p", class_ = "p-price-tag")
            paying_price = filter_string_num(p_price.find("span",class_="og-price").text)
            rating = filter_string_num(product_in.find("span", class_ = "rating-count").text)
            src = URL
            vendors = None
            img_src = None

            try:
                actual_price = filter_string_num(p_price.find("del").text)
            except:
                actual_price = None
            sql_value = (catagory,name,actual_price,paying_price,rating,src,vendors,img_src)
            try:
                sql_query = "INSERT INTO m1_data_points VALUES (?,?,?,?,?,?,?,?) "
                cursor.execute(sql_query,sql_value)
                conn.commit()
            except:
                pass
    except:
            pass
            
        
        


        
       
```
