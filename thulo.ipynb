```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import csv
```

```python
response = requests.get("https://smartdoko.com/")
soup = bs(response.content, 'html.parser')
```

```python
body = soup.body
```

```python
links_smartdoko = []
```

```python
links_smartdoko = []
links = body.find("div", class_ = "megadrop row")
a_tag = links.find_all("a", href = True)
for a in a_tag:
    links_smartdoko.append(a["href"])
```

```python
links_smartdoko
```

```python
for link in links_smartdoko[0:3]:
    try: 
        response = requests.get(link)
        soup = bs(response.content, 'html.parser')
    except:
        pass
    product_box = soup.body.find("div",id ="main-content")
    content_div = product_box.children
    

    
    
```

```python
for div in content_div:
    print(div)
```

<h4>Thulo.com</h4>

```python
response = requests.get("https://thulo.com/")
soup = bs(response.content, 'html.parser')
```

```python
menu_bar = soup.find_all("div", class_ = "ty-menu__submenu level-2")
```

```python
thulo_list = []
```

```python
for li in menu_bar:
    anchor = li.find_all('a',href= True)
    for a in anchor:
        thulo_list.append(a["href"])
```

```python
for a in thulo_list:
    response = requests.get(a)
    soup = bs(response.content, 'html.parser')
    print(soup.find("div",class_ = "grid-list et-grid-list"))
    
```
